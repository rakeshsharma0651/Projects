# Binary Functions

```python
Sp = pd.read_csv('5.2_Stockprice.csv',parse_dates=True,index_col=0)
Sp_s = Sp[['AAPL','MSFT','XOM']]
Sp_s = Sp_s.resample('B',fill_method='ffill')
#Calculating %age change
Spx_rets = Sp_s/Sp_s.shift(1) - 1
Spx_rets.head()
```

```python
#Other way to calculate %age Change
Apl = Sp_s['AAPL']
Returns = Apl.pct_change()
```

```python
corr = Spx_rets.rolling(126).corr(Returns) #126 is rolling window
```

```python
corr.plot()
```

```python
return_f = Sp_s.pct_change()
corr1=return_f.rolling(126).corr(return_f)
corr1.plot()
```

```python
From scipy.stats import percentileofscores
Score_2=lambda x : percentileofscores(x,0.02)
Result=pd.rolling_apply(Apl,126,Score_2)
Result.plot()
```
