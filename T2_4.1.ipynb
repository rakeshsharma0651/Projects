```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('4.1_tips.csv')
```

```python
df
```

```python
k1_means = df.groupby('day').mean().add_prefix('mean_')
```

```python
k1_means
```

```python
pd.merge(df,k1_means,left_on='day',right_index=True)
```

```python
key = ['smoker','day']
```

```python
df.groupby(key).mean()
```

```python
df.groupby(key).transform(np.mean)
```

```python
def top(df,n=5,column='total_bill'):
    return df.sort_values(by=column)[-n:]
df.groupby('smoker').apply(top)
```

```python
df.groupby(['smoker','day']).apply(top)
```

```python
df.pivot_table(index=['day','smoker'])
```

```python
df.pivot_table(['tip','size'],index=['time','day'],columns='smoker')
```

```python
#For Total, Margins used for all
df.pivot_table(['tip','size'],index=['time','day'],columns='smoker',margins=True)
```

```python
#For Total
df.pivot_table(['tip','size'],index=['time','day'],columns='smoker',aggfunc=len,margins=True)
```

```python
#For Total
df.pivot_table(['tip','size'],index=['time','day'],columns='smoker',aggfunc=sum,fill_value=0)
```
